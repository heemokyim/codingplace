```python
X = np.array([[2, 3, -2], [3, 5, 6], [2, 4, 3]])
print(X)
X = X.reshape(3, 3)
print(X)
```

```python
inv_X = np.linalg.inv(X)
print(inv_X)
```

```python
b = np.array([[-5], [12], [7]])
```

```python
x = np.dot(inv_X, b)
print(x)
```

```python
A_x = np.dot(X,x)
print(A_x)
print(b)
print(A_x - b)
# https://datascienceschool.net/view-notebook/148fc57f684c4dc48eeb5048ab0d45f2/
```

위와 같은 오차때문에 다음과 같은 계산은 피해야 한다
- 0.1 + 0.2 == 0.3

비교하려면 다음과 같이 해야 한다.
- round(0.1 + 0.2) = round(0.3)

왜 이런 현상이 나오나?
- 컴퓨터는 데이터를 2진형식으로 저장하기 때문에 그렇다
    - 언어의 문제가 아님

만약 행렬의 원소의 편차가 크면 역행렬과의 오차가 커진다

```python
from sklearn.datasets import load_boston
boston = load_boston()
X = boston.data
Y = boston.target

A = X[:, :]
B = Y[:]

w, resid, rank, s = np.linalg.lstsq(A, B)
residual = B - np.dot(A, w)
# print(np.linalg.norm(residual)**2)
# print(resid)
print(w)
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
X = boston.data
y = boston.target
A = X[:4, [0, 4, 5, 6]]  # 'CRIM', NOX', 'RM', 'AGE'
b = y[:4]
```

```python
Ainv = np.linalg.inv(A)

x = np.dot(Ainv, b)

print(x)
print(np.dot(A, x))
print(b)
```

```python
x, resid, rank, s = np.linalg.lstsq(A, b)
print(x)
```
