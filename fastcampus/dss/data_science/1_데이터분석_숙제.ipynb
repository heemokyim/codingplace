## 연습문제1
1. 기사 추천 = 수많은 뉴스 중 나에게 유익한 기사를 추려주는 추천엔진
2. 매출액 예상 = 식당 사장님 입장에서 다음달 매출액을 예측하고 싶은 경우
3. 해커 구분 시스템 = 해컨지 유전지 구분하는 시스템
    > 1. 각 24시간을 24-벡터로 표현함 (필요하다면 더 세분하게)
    > 2. 로그를 이용하여 유저가 특정시간대에 접속했다면 1, 아니면 0
    > 3. 유저가 가입한 후부터의 벡터들의 평균벡터를 구함
    > 4. 해커의 소행으로 보이는 수상한 날짜의 24-벡터와 이 평균벡터를 내적함
    > 5. 값이 높으면 유저일 확률이 높고 값이 낮으면 해커일 확률이 높다

## 연습문제2
#### 기사 추천
입력변수 = 기사의 분야, 독자의 관심분야, 조회수 & 출력변수 = 기사를 추천할지 아닌지를 이진값으로 나타낸 n-벡터
    - 뉴스분야 = 카테고리 데이터 
        > 경제, 사회, IT, 과학, 스포츠, 정치 etc
    - 관심분야 = 카테고리 데이터
    - 조회수 = 연속형 데이터

#### 매출액 예상
입력변수 = 날짜, 휴일유무, 날씨, 과거매출액, 원가, 인건비 & 출력변수 = 예상매출액
    - 날짜 = 스트링..? (이 부분은 시계열분석같은데 아직 잘 모르겠습니다)
    - 휴일유무 = 휴일인지 아닌지를 나타내는 이진값
        > 휴일이다 = 1 휴일이 아니다 = 0
    - 날씨 = 날씨가 좋다, 안좋다(비, 눈)의 이진값
    - 과거매출액, 원가, 인건비 = 연속형 데이터

#### 해커 구분 시스템
입력변수 = 유저로그 & 출력변수 = 유사도 스칼라
    - 유저로그 = ID, 활동시간, 활동명을 적은 스트링
    - 유저로그가 전처리를 거쳐 24-벡터로 변경되면 이걸 바탕으로 유사도 스칼라를 구함

## 연습문제5
#### 기사추천
기사의 분야와 나의 관심분야가 카테고리 값인데 이런 카테고리 값을 정량적으로 벡터화시켜야 예측에 용이할 것 같습니다. 연습문제 2번에 출력변수로 '추천할지 안할지 이진값으로 구분하는 n-벡터'라고 했는데 출력값을 n-차원 벡터가 나오도록 생각해놨었는데 구체적으로 어떻게 해야할 진 아직 모르지만 확실히 input 데이터에서부터 원하는 출력이 나오도록 다듬어야 한다는 걸 느꼈습니다.

#### 매출액 예상
매출액을 예상한다는 개념 자체가 시간적인 데이터도 다뤄야 하기 때문에 날짜 데이터와 관련한 분석의 필요성을 느꼈습니다. 단순히 알고리즘적인 문제해결이 아니라 시간을 고려한 수치적인 방법이 있어야 합니다. 또한 입출력변수를 설정하면서 '이거 말고도 뭔가 더 있을텐데'라며 고민해봤지만 도메인 지식이 부족해서 별다른 통찰을 얻지 못했습니다. 분석 기술의 필요성과 동시에 도메인 지식의 중요성을 깨달았습니다.

#### 해커 구분 시스템
사실 이 문제는 저번 시간에 배운 내적을 통한 유사도에 착안하여 만들었습니다. 유저 로그를 수집하여 이를 바탕으로 24-벡터를 만들고 연습문제 1번에 설명한대로 처리합니다.
