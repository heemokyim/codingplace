```python
b = 1
a = 0

# 1-1. 부정적분
x = sympy.symbols('x')
f = 3 * x**2 - 6 * x + 1

F = sympy.integrate(f)
Fb = F.subs(x, b)
Fa = F.subs(x, a)
print((Fb - Fa).evalf())
# -1.00000000000000

# 1-2. 수치적분
def f(x):
    return 3 * x**2 - 6 * x + 1

print(sp.integrate.quad(f, a, b))
# (-1.0, 1.3085085171449517e-14)
```

```python
b = 10
a = 1

# 2-1. 부정적분
x = sympy.symbols('x')
f = 2 + 6*x + 4*sympy.exp(x)+5*x**(-1)

F = sympy.integrate(f)
Fb = F.subs(x, b)
Fa = F.subs(x, a)
print((Fb - Fa).evalf())
# 88421.5029773780

# 2-2. 수치적분
def f(x):
    return 2 + 6*x + 4*np.exp(x)+5*x**(-1)

print(sp.integrate.quad(f, a, b))
# (88421.50297737827, 1.5276890734473408e-06)
```
