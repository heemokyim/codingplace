```python
df = pd.DataFrame(columns=['Name','Point'])
df['Name']=['A','B','C','D','E']
df['Point']=[random.randint(60,100) for each in range(5)]
df
```

```python
# 학점을 나타내는 컬럼을 추가해주세요
df['Grade'] = ''


def grade(score):
    tmp = score // 10
    if tmp==10:
        return 'A+'
    if tmp == 9:
        return 'A'
    elif tmp == 8:
        return 'B'
    elif tmp == 7:
        return 'C'
    else:
        return 'F'


df['Grade'] = df['Point'].apply(grade)
df
```

```python
# make data functions
# 사람 이름과 나이가 들어간 데이터를 만들기

import random, string

def get_name():
    names = ['Adam', 'Alan', 'Alex', 'Alvin', 'Andrew', 'Anthony', 'Arnold', 'Baldy', 'Baron', 'Billy', 'Boris', 'Bruno', 'Caley', 'Champ', 'Charlie', 'Clark']
    return random.choice(names)

def get_age(start=20, end=40):
    return random.randint(start, end)

def make_data(rows=10):
    datas = []
    for _ in range(rows):
        data = {"Age":get_age(), "Name":get_name()}
        datas.append(data)
    return datas
```

```python
data1 = make_data()
df1 = pd.DataFrame(data1)
df1

data2 = make_data()
df2 = pd.DataFrame(data2)
df2

# append
# df1에 df2 데이터를 추가하기
df3 = df1.append(df2)
df3

df3.reset_index(drop=True, inplace=True)
df3
```
