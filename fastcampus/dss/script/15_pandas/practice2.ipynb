```python
number = 16
names = ['Adam', 'Alan', 'Alex', 'Alvin', 'Andrew', 'Anthony', 'Arnold', 'Baldy', 
         'Baron', 'Billy', 'Boris', 'Bruno', 'Caley', 'Champ', 'Charlie', 'Clark']

df = pd.DataFrame(columns=["Name","Math","Science","English"])

# set data
def set_data():
    return list(np.random.randint(60,100,(number,3)))

# set total

# set grades
def get_grade(point):
    if point >= 270:
        return "A"
    elif point > 240:
        return "B"
    elif point >= 210:
        return "C"
    elif point >= 180:
        return "D"
    else:
        return "F"
    
# set sort
df['Name']=names

# for _ in range(1,3+1):
#     df.iloc[:,_] =set_data()

df.iloc[:,1:4]=set_data()

df['Total'] = df['Math']+df['Science']+df['English']

df['Grade'] = df['Total'].apply(get_grade)

df.sort_values('Grade',inplace=True)

df
```

```python
def how_much(Amount):
    title = ''
    unit = 1E5
    if Amount > 8 * unit:
        title += 'Rich'
    elif Amount > 7 * unit:
        title += 'Middle-wealth'
    elif Amount > 6 * unit:
        title += 'Not needy at least'
    else:
        title += 'Aweful'
    return title

df['Status'] = df['Amount'].apply(how_much)
df
```
