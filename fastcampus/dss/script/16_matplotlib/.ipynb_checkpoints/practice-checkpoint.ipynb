```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
data = np.random.randint(10, size=10)
opts = {'c':'r','ls':'--'}
plt.plot(data,**opts)

plt.show()
```

```python
data = np.random.randint(10, size=10)
xtick = range(0, 100, 10)
plt.plot(xtick,data, c="r", lw=4, ls="--", marker="D", ms="10", mec="g", mew="3", mfc="w")
plt.show()
```

```python
data = np.random.randint(10, size=10)
xtick = range(0, 100, 10)
plt.plot(xtick,data, c="r", lw=4, ls="--", marker="D", ms="10", mec="g", mew="3", mfc="w")

plt.grid(True)

plt.show()
```

```python
data1 = np.random.randint(10, size=10)
data2 = np.random.randint(10, size=10)

xtick = range(0, 100, 10)

plt.plot(data1, label='data1')
plt.plot(data2, label='data2')

plt.legend(loc=0)
plt.title('Title')

plt.show()
```

```python
data1 = np.random.randint(5, size=5)
data2 = np.random.randint(5, size=5)

plt.figure(figsize=(5,5))
plt.subplot(3, 1, 1) # 2*1에서 1번 위치
plt.plot(data1)

plt.subplot(3, 1, 3) # 2*1에서 2번 위치
plt.plot(data2)

plt.figure(figsize=(5,5))
plt.plot(data1)

plt.show()
```

```python
y = np.random.randint(5, 10, size=5)
x = range(5) 
error = np.random.randint(1, 5, size=5)
print(y, error)
plt.bar(x, y, yerr=error, align='center', alpha=0.2)
plt.show()
```

```python
plt.barh(x,y,xerr=error, align='center',alpha=0.9)
plt.show()
```

```python
y1 = np.random.randint(5, 10, size=5)
y2 = np.random.randint(1, 5, size=5)
print(y1, y2)
plt.subplots()
x = range(5)
plt.bar(x, y1)
plt.bar(x, y2)
plt.tight_layout()
plt.show()
```

```python
y1 = np.random.randint(1, 5, size=5)
y2 = np.random.randint(1, 5, size=5)
bar_width, opacity = 0.4, 0.9
print(y1, y2, bar_width, opacity)
x = np.arange(5)

plt.bar(x, y1, bar_width, alpha=opacity, color='b', label='randint_1', align="edge")
plt.bar(x + bar_width, y2, bar_width, alpha=opacity, color='r', label='randint_2', align="edge")

plt.xlabel('X Label') # X 라벨
plt.ylabel('Value')   # Y 라벨
plt.xticks(x + bar_width, ('A', 'B', 'C', 'D', 'E')) # xtick 추가
plt.title('Random Integer') # 제목추가
# plt.legend() # 범례추가
plt.show()
```

```python
y1 = np.random.randint(1, 5, size=5)
y2 = np.random.randint(1, 5, size=5)
bar_width, opacity = 0.4, 0.9
print(y1, y2, bar_width, opacity)
x = np.arange(5)

plt.bar(x, y1, bar_width, alpha=opacity, color='b', label='randint_1', align="edge")
plt.bar(x + bar_width, y2, bar_width, alpha=opacity, color='r', label='randint_2', align="edge")

plt.xlabel('X Label') # X 라벨
plt.ylabel('Value')   # Y 라벨
plt.xticks(x + bar_width, ('A', 'B', 'C', 'D', 'E')) # xtick 추가
plt.title('Random Integer') # 제목추가
plt.legend() # 범례추가
plt.show()
```
