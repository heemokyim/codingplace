```python
pivot1 = tips.groupby(['time','sex']).agg('sum').reset_index()
pivot1
```

```python
pivot1 = pivot1.dropna(how='any',axis=0)
pivot1
```

```python
pivot2 = pivot1.pivot('time','sex','tip')
pivot2
```

```python
sns.heatmap(pivot2, cmap=sns.light_palette("purple", as_cmap=True), annot=True, fmt=".2f")
```
