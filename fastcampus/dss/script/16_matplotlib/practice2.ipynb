```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def sinplot():
    x = np.linspace(0, 10, 101) # 0-10 101 등분
    for i in range(6):
        plt.plot(x, np.sin(i + x)) # 60도씩 sin이 이동된 데이터를 그려줌
sinplot()
```

```python
sns.set()
sns.set_style('whitegrid')
sinplot()
```
