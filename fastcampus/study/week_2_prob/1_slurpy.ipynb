```python
import re
```

```python
def check_slurpy(Slurpy):
    '''
    1. Slurpy를 입력받음
    2. 알파벳 C,H를 기준으로 Slimp, Slump를 나눔
    '''

    C = Slurpy.rfind('C')
    H = Slurpy.rfind('H')

    idx = max(C,H) if C*H<0 else min(C,H)
    
    Slimp, Slump = Slurpy[:idx+1], Slurpy[idx+1:]

    if check_slimp(Slimp):
        if check_slump(Slump):
            return True
    return False
```

```python
def check_slump(sample):
    '''
    1. 슬럼프 규칙에 맞는 지 검사함
    2. 검사된 애들은 빼는 방식
    3. 검사하다가 위배되는 애들이 있으면 False를 리턴해서 구분함
    '''

    while len(sample) != 0:
        first = sample[0]
        # 첫글자가 D와 E가 아니면 슬럼프가 아님
        if first != 'D' and first != 'E' and first != 'G':
            return False

        # 첫글자가 G인 경우,
        elif first == 'G':
            # 문자열의 길이가 1이면 슬럼프 (끝에 왔다는 뜻)
            if len(sample) == 1:
                return True
            # 문자열의 길이가 1이 아니면 슬럼프가 아님 (규칙에 위배됨)
            else:
                return False

        # 만약 D나 E면 첫번째글자 삭제
        else:
            sample = sample.replace(first, '')

        # 하나 이상 반복되는 F들을 찾음
        F = re.findall('F+', sample)
        # 있으면 지움
        if F:
            sample = sample.replace(F[0], '', 1)
        # 없으면 슬럼프가 아님 (규칙에 위배됨)
        else:
            return False
    return False
```

```python
def check_slimp(sample):
    '''
    1. 작동방식은 check_slump와 같음
    2. 슬림프의 앞글자와 뒷글자를 검사함
    3. 두 개의 문자 슬림프,세 개 이상의 슬림프인 경우로 나눠서 검사함
    '''

    while len(sample) != 0:
        # 1. 앞글자 뒷글자 검사
        first = sample[0]
        end = sample[-1]

        # 2. 두 개의 문자 슬림프
        if first == 'A' and end == 'H' and len(sample) == 2:
            return True

        # 2. 세 개 이상의 슬림프
        elif first == 'A' and end == 'C':
            second = sample[1]

            # 2-1. AB + 슬림프 + C
            if second == 'B' and len(sample)!=3:
                # 앞에 A,B 뒤에 C를 지움
                listed_sample = list(sample)

                listed_sample[0] = ""
                listed_sample[1] = ""
                listed_sample[-1] = ""

                sample = "".join(listed_sample)

            # 2-2. A + 슬럼프 + C
            else:
                # 앞에 A, 뒤에 C를 지우고 check_slump
                listed_sample = list(sample)
                listed_sample[0] = ""
                listed_sample[-1] = ""
                trimmed_sample = "".join(listed_sample)

                if check_slump(trimmed_sample):
                    return True
                else:
                    return False
        # 두 개의 슬림프, 세 개 이상의 슬림프인 경우에 해당하지 않으면
        # 슬림프가 아니므로 return False
        else:
            return False
    #. 만약 while문에서 False가 한 번도 안걸리고 빠져나오면 slimp임
    return True
```

```python
Slumps = ['DFG', 'EFG', 'DFFFFFG', 'DFDFDFDFG', 'DFEFFFFFG']
n_Slumps = ['DFEFF', 'EFAHG', 'DEFG', 'DG', 'EFFFFDG']

Slimps=['AH', 'ABAHC', 'ABABAHCC', 'ADFGC', 'ADFFFFGC', 'ABAEFGCC', 'ADFDFGC']
n_Slimps=['ABC', 'ABAH', 'DFGC', 'ABABAHC', 'SLIMP', 'ADGC']

Slurpys =['AHDFG', 'ADFGCDFFFFFG', 'ABAEFGCCDFEFFFFFG']
n_Slurpys=['AHDFGA', 'DFGAH', 'ABABCC']
```

```python
for yes, no in zip(Slumps, n_Slumps):
    a = check_slump(yes)
    b = check_slump(no)
    print(a, b)
print()

for yes, no in zip(Slimps, n_Slimps):
    a = check_slimp(yes)
    b = check_slimp(no)
    print(a, b)
print()

for yes, no in zip(Slurpys, n_Slurpys):
    a = check_slurpy(yes)
    b = check_slurpy(no)
    print(a, b)
print()
```
