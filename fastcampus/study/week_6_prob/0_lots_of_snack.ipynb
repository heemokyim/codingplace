```python
def eatCookie(cookies):
    dp = [1 for i in range(len(cookies))]
    
    print(cookies)
    
    for i in range(1,len(cookies)):
        for j in range(0,i):
            if cookies[i] > cookies[j]:
                if dp[i] < dp[j] + 1 :
                    dp[i] = dp[j] + 1
        print(dp)
    return max(dp)
# 아래는 테스트로 출력해 보기 위한 코드입니다.
print(eatCookie([1, 4, 2, 6, 3, 4, 1, 5]))
```

```python
def lots_cookies(cookies):
    count = 0
    
    while cookies:
        cookies = eatCookie(cookies)
        count += 1
    return count
```
