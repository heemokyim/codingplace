```python
from pyspark import SparkConf, SparkContext

conf = SparkConf().setMaster("local").setAppName("WordCount")
sc = SparkContext(conf = conf)

lines = sc.textFile("file:/home/ej/github/codingplace/spark/dataset/Book.txt")
```

```python
lines.map(lambda x:x).collect()
```

```python
each_word = lines.flatMap(lambda x:x.split())
```

```python
each_word.collect()[:10]
```

```python
result_dict = each_word.countByValue()
```

```python
import collections
for k,v in collections.OrderedDict(result_dict).items():
    print(k,v)
```
